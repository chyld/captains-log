```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(suppress=True)
%matplotlib inline
```

```python
x = np.linspace(-5, 0).reshape(50,1)
```

```python
def generate_sample(x):
    return np.array([np.sin(n) + np.random.normal(10, 0.1) for n in x])
```

```python
y = generate_sample(x)
```

```python
plt.scatter(x, y);
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras import optimizers
```

```python
model = Sequential()
model.add(Dense(units=50, activation='relu', input_shape=(1,)))
model.add(Dense(units=5, activation='relu'))
model.add(Dense(units=1, activation='linear'))

sgd = optimizers.SGD(lr=0.01)
model.compile(loss='mse', optimizer=sgd, metrics=['mse'])
model.summary()
```

```python
x.shape
```

```python
history = model.fit(x, y, epochs=200, batch_size=10, verbose=2)
```

```python
y_hat = model.predict(x).reshape(1,50)
```

```python
plt.scatter(x,y);
plt.scatter(x,y_hat);
```
