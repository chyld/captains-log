```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.set_printoptions(suppress=True)
x = np.arange(-5, 5)
y = 3*x + np.random.normal(0, .5)
```

```python
plt.scatter(x,y);
```

```python
m = 2
print(x)
print(y)
print(m*x)
```

```python
# squared residuals
losses = (y-(m*x)) ** 2
losses
```

```python
# sum of squared residuals
losses.sum()
```

```python
plt.scatter(x, losses);
```
